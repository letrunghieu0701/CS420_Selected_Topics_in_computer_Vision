```python
import matplotlib as plt
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import WhitespaceTokenizer 
```

```python
stop_words = stopwords.words('english')

white_space_tokenizer = WhitespaceTokenizer()
```

```python
def preprocessString(string_data):
    # Lower the text
    preprocess_data = string_data.lower()
        
    # Tokenize word by white space
    preprocess_data = white_space_tokenizer.tokenize(preprocess_data)
    
    # Remove punctuations, each punctuation = space, ex: ""information @#$retrieval" -> "information    retrieval"
    preprocess_data = re.sub('[%s]' % re.escape(string.punctuation), ' ', preprocess_data)   
    
    # Remove stop words
    preprocess_data = [word for word in preprocess_data if word not in stop_words]
        
    return preprocess_data
```

```python
file_names = ['Day_1_29-10-2021.txt', 'Day_2_30-10-2021.txt', 'Day_3_31-10-2021.txt', 'Day_4_01-11-2021.txt']
```

```python
for file_name in file_names:
    with open(file_name, "r") as f:
        content = f.readlines()
        f.close()

        paper_names_tokenzied = list()
        for i in range(0, len(content), 4):
            paper_name_preprocessed = preprocessString(content[i])
            paper_names_tokenzied.append(paper_name_preprocessed)          
        

        author_names_tokenzied = list()
        for i in range(1, len(content), 4):
            author_name_preprocessed = preprocessString(content[i])
            author_names_tokenzied.append(author_name_preprocessed)      
```
