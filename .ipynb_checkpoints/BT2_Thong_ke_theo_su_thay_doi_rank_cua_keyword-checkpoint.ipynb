**Import libraries**

```python
import numpy as np
import nltk
# Use below line to download stopword
# nltk.download('stopwords')
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

**Stopword and file names**

```python
stop_words = stopwords.words('english')
file_names = ['Day_1_29-10-2021.txt', 'Day_2_30-10-2021.txt', 'Day_3_31-10-2021.txt', 'Day_4_01-11-2021.txt']
```

**Reading content of text files**

```python
paper_names = list()
author_names = list()

for file_name in file_names:
    with open(f"Texts\\{file_name}", "r") as f:
        content = f.readlines()
        f.close()
        
    for i in range(0, len(content), 4):
        paper_names.append(content[i])
        
    for i in range(1, len(content), 4):
        author_names.append(content[i])
```

**Special tokenizer and function to statistic terms**

```python
def my_tokenizer(text):
    """This function is used to deal with author names"""
    # Remove space after comma and newline charecter, after that, split based on comma
    return text.replace(', ', ',').replace('\n', '').split(',')
        
def statistic_keyword(collection_of_text, type_of_data, graph_name):
    # statistic term
    if type_of_data == "paper_name":
        vectorizer = CountVectorizer(lowercase=True, stop_words=stop_words)
        y_label = "Keyword"
    elif type_of_data == "author_name":
        vectorizer = CountVectorizer(lowercase=False, stop_words=stop_words, tokenizer=my_tokenizer)
        y_label = "Author name"
        
    count = sum((vectorizer.fit_transform(collection_of_text)).toarray())
    x = vectorizer.get_feature_names()
    
    # Print some information
    print("Total term: ",len(count))
    unique, counts = np.unique(count, return_counts=True)
    print("Frequency\t Number of term have this frequency")
    print(np.asarray((unique, counts)).T)

    count,x = zip(*sorted(zip(count,x), reverse=True))
    
    # Draw graph
    plt.figure(figsize = (25, len(count)//40))
    plt.barh(x, count, color ='maroon', align='edge', height=0.8)

    plt.yticks(fontsize=1)
    plt.xlabel("Frequency")
    plt.ylabel(y_label)
    plt.title(f"Frequency of {y_label} in ICCV 2019 conference")
    plt.savefig(f"Result_images\\{graph_name}.pdf")
    plt.show()
```

```python
statistic_keyword(paper_names, "paper_name", "Keyword_frequency")
statistic_keyword(author_names, "author_name", "Author_name_frequency")
```
